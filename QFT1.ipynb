```python
import time
import numpy as np
import pennylane as qml
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset
from skimage.metrics import structural_similarity

# -----------------------------
# Reproducibility + CPU stability
# -----------------------------
SEED = 0
torch.manual_seed(SEED)
np.random.seed(SEED)
torch.set_num_threads(1)

DTYPE = torch.float32  # <<< enforce everywhere

# -----------------------------
# Config
# -----------------------------
IMG_SIZE = 16
TRAIN_SUBSET = 1024
TEST_SUBSET = 256
BATCH_SIZE = 2
EPOCHS = 2
MAX_TRAIN_BATCHES = 80
DEVICE = torch.device("cpu")

# Quantum settings
FRQI_K = None
DIFF_METHOD = "parameter-shift"

# Optimization
LR = 2e-4
CLIP_NORM = 20.0

# Loss weights (scheduled)
W_EDGE = 0
W_TV   = 0.05
W_BG   = 10.0
BG_THRESH = 0.12

# Logging
PREVIEW_EVERY = 20
PRINT_EACH_IMAGE = False

# -----------------------------
# Schedules
# -----------------------------
SIGMA_RANGES = [
    (0.12, 0.20),
    (0.08, 0.16),
    (0.05, 0.12),
]
W_IMG_SCHED = [80.0, 120.0, 200.0]
W_RES_SCHED = [400.0, 300.0, 200.0]
RES_SCALE_SCHED = [2.0, 1.5, 1.0]

def get_sched(arr, epoch_idx):
    return arr[epoch_idx] if epoch_idx < len(arr) else arr[-1]

def get_sigma_range(epoch_idx):
    return SIGMA_RANGES[epoch_idx] if epoch_idx < len(SIGMA_RANGES) else SIGMA_RANGES[-1]

# -----------------------------
# Data
# -----------------------------
transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE), interpolation=transforms.InterpolationMode.BILINEAR),
    transforms.ToTensor(),  # float32 in [0,1]
])

trainset_full = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
testset_full  = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

def make_subset(ds, n, seed):
    if n is None:
        return ds
    g = torch.Generator().manual_seed(seed)
    idx = torch.randperm(len(ds), generator=g)[:n]
    return Subset(ds, idx)

trainset = make_subset(trainset_full, TRAIN_SUBSET, seed=SEED)
testset  = make_subset(testset_full, TEST_SUBSET,  seed=SEED + 1)

train_loader = DataLoader(trainset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
test_loader  = DataLoader(testset, batch_size=BATCH_SIZE, shuffle=False, drop_last=False)

print(f"Train size: {len(trainset)} | Test size: {len(testset)} | IMG_SIZE={IMG_SIZE} | BATCH_SIZE={BATCH_SIZE}")

# -----------------------------
# Helpers
# -----------------------------
@torch.no_grad()
def add_noise_sigma(clean, sigma_float):
    # clean: (B,1,H,W) float32
    sigma = torch.tensor(float(sigma_float), dtype=DTYPE, device=clean.device)
    noisy = clean + torch.randn_like(clean, dtype=DTYPE) * sigma
    return noisy.clamp(0.0, 1.0)

def show_image(img2d, title=None):
    if isinstance(img2d, torch.Tensor):
        img2d = img2d.detach().cpu().numpy()
    plt.imshow(img2d, cmap="gray", vmin=0.0, vmax=1.0)
    if title is not None:
        plt.title(title)
    plt.axis("off")
    plt.show()

def psnr_torch(ref, out, eps=1e-8):
    ref = ref.to(dtype=DTYPE)
    out = out.to(dtype=DTYPE)
    mse = torch.mean((ref - out) ** 2)
    return 10.0 * torch.log10(torch.tensor(1.0, dtype=DTYPE, device=ref.device) / (mse + eps))

def ssim_np(ref, out):
    ref = ref.detach().cpu().numpy()
    out = out.detach().cpu().numpy()
    return structural_similarity(ref, out, data_range=1.0)

def total_variation(x):
    return (x[:, :, 1:] - x[:, :, :-1]).abs().mean() + (x[:, 1:, :] - x[:, :-1, :]).abs().mean()

# Sobel kernels float32
_sobel_x = torch.tensor([[[-1, 0, 1],
                          [-2, 0, 2],
                          [-1, 0, 1]]], dtype=DTYPE).unsqueeze(0)
_sobel_y = torch.tensor([[[-1, -2, -1],
                          [ 0,  0,  0],
                          [ 1,  2,  1]]], dtype=DTYPE).unsqueeze(0)

def edge_map(x):  # x: (B,H,W)
    x = x.to(dtype=DTYPE).unsqueeze(1)
    gx = F.conv2d(x, _sobel_x.to(x.device, dtype=x.dtype), padding=1)
    gy = F.conv2d(x, _sobel_y.to(x.device, dtype=x.dtype), padding=1)
    return torch.sqrt(gx * gx + gy * gy + 1e-8).squeeze(1)

# -----------------------------
# Quantum Generator
# -----------------------------
class QuantumGenerator:
    def __init__(self, num_pos, depth, frqi_K=None):
        self.num_pos = int(num_pos)
        self.wires = self.num_pos + 1
        self.depth = int(depth)
        self.frqi_K = frqi_K

        # shots untouched
        self.dev = qml.device("lightning.qubit", wires=self.wires, shots=None)

        self.qnode = qml.QNode(
            self._circuit,
            self.dev,
            interface="torch",
            diff_method=DIFF_METHOD,
        )

        self._flip_masks = []
        for idx in range(2**self.num_pos):
            bin_str = format(idx, f"0{self.num_pos}b")
            flips = [j for j, bit in enumerate(bin_str) if bit == "0"]
            self._flip_masks.append(flips)

    @staticmethod
    def frqi_angles(image_2d):
        image_2d = image_2d.to(dtype=DTYPE)
        return image_2d.flatten() * (torch.pi / 2.0)

    def _apply_frqi(self, angles):
        pos_wires = list(range(self.num_pos))
        color_wire = self.num_pos

        num_pixels = int(angles.shape[0])
        if self.frqi_K is not None:
            num_pixels = min(num_pixels, int(self.frqi_K))

        for idx in range(num_pixels):
            theta = angles[idx]
            flips = self._flip_masks[idx]

            for j in flips:
                qml.PauliX(pos_wires[j])

            qml.ctrl(qml.RY, control=pos_wires)(2.0 * theta, wires=color_wire)

            for j in flips:
                qml.PauliX(pos_wires[j])

    def _frqi(self, image_2d):
        for w in range(self.num_pos):
            qml.Hadamard(wires=w)
        angles = self.frqi_angles(image_2d)
        self._apply_frqi(angles)

    def _hea(self, weights):
        for l in range(self.depth):
            for q in range(self.num_pos + 1):
                tx = torch.pi * torch.tanh(weights[l, q, 0])
                ty = torch.pi * torch.tanh(weights[l, q, 1])
                tz = torch.pi * torch.tanh(weights[l, q, 2])
                qml.RX(tx, wires=q)
                qml.RY(ty, wires=q)
                qml.RZ(tz, wires=q)

            for q in range(self.num_pos):
                qml.CNOT(wires=[q, q + 1])

    def _circuit(self, image_2d, weights):
        self._frqi(image_2d)
        qml.QFT(wires=range(self.num_pos))
        self._hea(weights)
        qml.adjoint(qml.QFT)(wires=range(self.num_pos))
        return qml.probs(wires=range(self.num_pos + 1))

    def qgen(self, image_2d, weights):
        return self.qnode(image_2d, weights)

    def reconstruct_image_torch(self, probs, img_shape, eps=1e-8):
        probs = probs.to(dtype=DTYPE)
        probs2 = probs.view(2**self.num_pos, 2)
        p0 = probs2[:, 0]
        p1 = probs2[:, 1]
        img = p1 / (p0 + p1 + eps)
        return img.view(*img_shape)

# -----------------------------
# Residual Head
# -----------------------------
class ResidualHead(nn.Module):
    def __init__(self, channels=32):
        super().__init__()
        self.feat = nn.Sequential(
            nn.Conv2d(2, channels, 3, padding=1),
            nn.GroupNorm(4, channels),
            nn.SiLU(),
            nn.Conv2d(channels, channels, 3, padding=1),
            nn.GroupNorm(4, channels),
            nn.SiLU(),
            nn.Conv2d(channels, channels, 3, padding=1),
            nn.GroupNorm(4, channels),
            nn.SiLU(),
        )
        self.gate_out = nn.Conv2d(channels, 1, 1)
        self.res_out  = nn.Conv2d(channels, 1, 1)

        # Start gate CLOSED (prevents copying noisy)
        nn.init.zeros_(self.gate_out.weight)
        nn.init.constant_(self.gate_out.bias, -2.0)

        nn.init.normal_(self.res_out.weight, mean=0.0, std=0.01)
        nn.init.zeros_(self.res_out.bias)

    def forward(self, noisy, qout):
        # enforce float32 here (fixes your crash)
        noisy = noisy.to(dtype=DTYPE)
        qout  = qout.to(dtype=DTYPE)
        x = torch.stack([noisy, qout], dim=1)  # (B,2,H,W)
        x = x.to(dtype=DTYPE)
        f = self.feat(x)
        gate = torch.sigmoid(self.gate_out(f)).squeeze(1)
        res  = torch.tanh(self.res_out(f)).squeeze(1)
        return gate, res

# -----------------------------
# Setup quantum sizes
# -----------------------------
H = W = IMG_SIZE
num_pos = int(2 * np.ceil(np.log2(H)))            # 16 -> 8
wires = num_pos + 1
#depth = max(2, int(np.ceil(np.log2(num_pos))))    # 16 -> 3
depth=2
QG = QuantumGenerator(num_pos=num_pos, depth=depth, frqi_K=FRQI_K)
print(f"num_pos={num_pos} wires={wires} depth={depth} shots={QG.dev.shots} FRQI_K={FRQI_K}")

gen_weights = nn.Parameter(0.01 * torch.randn(depth, wires, 3, dtype=DTYPE, device=DEVICE))
head = ResidualHead(channels=32).to(DEVICE).to(dtype=DTYPE)

opt = optim.Adam([gen_weights] + list(head.parameters()), lr=LR)

# -----------------------------
# Warmup compile + SHOW circuit
# -----------------------------
with torch.no_grad():
    dummy = torch.zeros((H, W), dtype=DTYPE, device=DEVICE)
    _ = QG.qgen(dummy, gen_weights)
print("[WARMUP] qnode compiled (quantum is active).")

# Print a readable circuit once
try:
    print("\n[QUANTUM CIRCUIT - TEXT]")
    print(QG.qnode.tape)  # operations list
except Exception as e:
    print(f"[WARN] Could not print qnode.tape: {e}")

try:
    print("\n[QUANTUM CIRCUIT - DRAW]")
    drawer = qml.draw(QG.qnode, decimals=2, max_length=120)
    print(drawer(dummy, gen_weights))
except Exception as e:
    print(f"[WARN] Could not draw circuit: {e}")

_qnode_info_printed = False

def quantum_forward_batch(noisy_2d, res_scale):
    global _qnode_info_printed

    # force dtype float32 to avoid your "double vs float" crash
    noisy_2d = noisy_2d.to(device=DEVICE, dtype=DTYPE)

    q_list = []
    for i in range(noisy_2d.size(0)):
        probs = QG.qgen(noisy_2d[i], gen_weights)  # quantum call
        recon = QG.reconstruct_image_torch(probs, (H, W))
        q_list.append(recon)

        if (not _qnode_info_printed) and (i == 0):
            try:
                ops_n = len(QG.qnode.tape.operations)
                print(f"    [INFO] quantum ops in tape: {ops_n} | shots={QG.dev.shots}")
            except Exception:
                pass
            _qnode_info_printed = True

    q_out = torch.stack(q_list, dim=0).clamp(0.0, 1.0).to(dtype=DTYPE)  # (B,H,W)

    gate, res = head(noisy_2d, q_out)

    # remove drift per-image
    res = res - res.mean(dim=(1, 2), keepdim=True)

    # schedule residual strength
    res = res_scale * res

    pred_res = gate * res
    fake = (noisy_2d + pred_res).clamp(0.0, 1.0)

    return q_out, gate, res, pred_res, fake

@torch.no_grad()
def evaluate_fixed_sigma(loader, sigma=0.10, res_scale=1.0):
    base_psnrs, psnrs, ssims, deltas = [], [], [], []
    for clean, _ in loader:
        clean = clean.to(DEVICE, dtype=DTYPE)
        noisy = add_noise_sigma(clean, float(sigma))

        clean_2d = clean.squeeze(1).to(dtype=DTYPE)
        noisy_2d = noisy.squeeze(1).to(dtype=DTYPE)

        _, _, _, _, fake = quantum_forward_batch(noisy_2d, res_scale=res_scale)

        for i in range(fake.size(0)):
            b = psnr_torch(clean_2d[i], noisy_2d[i]).item()
            p = psnr_torch(clean_2d[i], fake[i]).item()
            s = ssim_np(clean_2d[i], fake[i])
            base_psnrs.append(b)
            psnrs.append(p)
            ssims.append(s)
            deltas.append(p - b)

    return float(np.mean(base_psnrs)), float(np.mean(psnrs)), float(np.mean(ssims)), float(np.mean(deltas))
import lpips
lpips_fn = lpips.LPIPS(net="alex").to(DEVICE)
lpips_fn.eval()
# from DISTS_pytorch import DISTS
# dists_fn = DISTS().to(DEVICE)
# dists_fn.eval()
@torch.no_grad()
def lpips_score(x, y, size=64):
    """
    x, y: (H, W) in [0,1]
    Upsamples to size×size before LPIPS
    """
    x = x.unsqueeze(0).unsqueeze(0)  # (1,1,H,W)
    y = y.unsqueeze(0).unsqueeze(0)

    x = F.interpolate(x, size=(size, size), mode="bilinear", align_corners=False)
    y = F.interpolate(y, size=(size, size), mode="bilinear", align_corners=False)

    x = x.repeat(1, 3, 1, 1)
    y = y.repeat(1, 3, 1, 1)

    x = x * 2 - 1
    y = y * 2 - 1

    return lpips_fn(x, y).item()

# @torch.no_grad()
# def dists_score(x, y):
#     x = x.unsqueeze(0).repeat(1,3,1,1)
#     y = y.unsqueeze(0).repeat(1,3,1,1)
#     return dists_fn(x, y).item()

# -----------------------------
# Training loop
# -----------------------------
global_t0 = time.time()
batch_history = {
    "loss": [],
    "edge": [],
    "lpips": [],
    "dists": []
}

for epoch in range(EPOCHS):
    W_IMG = get_sched(W_IMG_SCHED, epoch)
    W_RES = get_sched(W_RES_SCHED, epoch)
    RES_SCALE = get_sched(RES_SCALE_SCHED, epoch)
    smin, smax = get_sigma_range(epoch)

    print(f"\n=== Epoch {epoch+1}/{EPOCHS} ===")
    print(f"  [SCHED] sigma~U({smin:.2f},{smax:.2f}) | W_IMG={W_IMG:.1f} W_RES={W_RES:.1f} | RES_SCALE={RES_SCALE:.2f}")

    epoch_t0 = time.time()

    for batch_idx, (clean, _) in enumerate(train_loader):
        if MAX_TRAIN_BATCHES is not None and batch_idx >= MAX_TRAIN_BATCHES:
            print(f"Stopping early after {MAX_TRAIN_BATCHES} train batches.")
            break

        batch_t0 = time.time()

        clean = clean.to(DEVICE, dtype=DTYPE)  # (B,1,H,W)

        # sigma sample explicitly float32 -> then float
        sigma = (smin + (smax - smin) * float(torch.rand(1, dtype=DTYPE).item()))
        noisy = add_noise_sigma(clean, sigma)

        clean_2d = clean.squeeze(1).to(dtype=DTYPE)
        noisy_2d = noisy.squeeze(1).to(dtype=DTYPE)

        print(f"\n[Epoch {epoch+1} | Batch {batch_idx}] start clean={tuple(clean_2d.shape)} sigma={sigma:.4f}")

        with torch.no_grad():
            base_psnr = float(np.mean([psnr_torch(clean_2d[i], noisy_2d[i]).item() for i in range(clean_2d.size(0))]))
        print(f"  [BASELINE] PSNR(noisy vs clean)={base_psnr:.2f} dB")

        # ---- STEP ----
        step_t0 = time.time()
        opt.zero_grad()

        q_out, gate, res, pred_res, fake = quantum_forward_batch(noisy_2d, res_scale=RES_SCALE)

        target_res = (clean_2d - noisy_2d)

        mse_img = ((fake - clean_2d) ** 2).mean()
        mse_res = ((pred_res - target_res) ** 2).mean()

        e_fake = edge_map(fake)
        e_clean = edge_map(clean_2d)
        edge_l = ((e_fake - e_clean) ** 2).mean()

        tv = total_variation(fake)

        bg_mask = (clean_2d < BG_THRESH).float()
        bg = (fake * bg_mask).mean()

        loss = W_IMG * mse_img + W_RES * mse_res + W_EDGE * edge_l + W_TV * tv + W_BG * bg
        loss.backward()

        torch.nn.utils.clip_grad_norm_([gen_weights] + list(head.parameters()), max_norm=CLIP_NORM)
        grad_norm = gen_weights.grad.norm().item() if gen_weights.grad is not None else float("nan")

        opt.step()

        gate_mean = gate.mean().item()
        res_abs_mean = res.abs().mean().item()
        w_mean = gen_weights.data.mean().item()
        w_std = gen_weights.data.std().item()

        batch_history["loss"].append(loss.item())
        batch_history["edge"].append(edge_l.item())
        print(
            f"  [STEP] loss={loss.item():.4f} | "
            f"MSEimg={mse_img.item():.6f} MSEres={mse_res.item():.6f} EDGE={edge_l.item():.6f} "
            f"TV={tv.item():.5f} BG={bg.item():.5f} | "
            f"gate_mean={gate_mean:.4f} res_abs_mean={res_abs_mean:.4f} | "
            f"grad_norm={grad_norm:.3e} | w_mean={w_mean:.3e} w_std={w_std:.3e} | "
            f"time={time.time()-step_t0:.2f}s"
        )
        lpips_vals = []
        with torch.no_grad():
            for i in range(fake.size(0)):
                lpips_vals.append(lpips_score(fake[i], clean_2d[i]))
        batch_lpips = np.mean(lpips_vals)
        batch_history["lpips"].append(batch_lpips)

        # ---- Metrics ----
        m_t0 = time.time()
        with torch.no_grad():
            psnrs = [psnr_torch(clean_2d[i], fake[i]).item() for i in range(fake.size(0))]
            ssims = [ssim_np(clean_2d[i], fake[i]) for i in range(fake.size(0))]
            deltas = [psnrs[i] - psnr_torch(clean_2d[i], noisy_2d[i]).item() for i in range(fake.size(0))]
        print(
            f"  [METRICS train] PSNR={float(np.mean(psnrs)):.2f} dB | "
            f"SSIM={float(np.mean(ssims)):.3f} | "
            f"DELTA={float(np.mean(deltas)):+.2f} dB | time={time.time()-m_t0:.2f}s"
        )

        print(f"[Epoch {epoch+1} | Batch {batch_idx}] done in {time.time()-batch_t0:.2f}s")

        if (batch_idx % 10) == 0:
            show_image(noisy_2d[0], title="Noisy input")
            show_image(fake[0], title="Denoised output")
            show_image(clean_2d[0], title="Clean target")
            show_image((fake[0] - noisy_2d[0]).abs(), title="|Output - Noisy|")

    # ---- Evaluation at 3 sigmas ----
    eval_t0 = time.time()
    for s_eval in [0.05, 0.10, 0.15]:
        bpsnr, mpsnr, mssim, dpsnr = evaluate_fixed_sigma(test_loader, sigma=s_eval, res_scale=RES_SCALE)
        print(
            f"[EVAL sigma={s_eval:.2f}] BaselinePSNR={bpsnr:.2f} dB | "
            f"ModelPSNR={mpsnr:.2f} dB | DELTA={dpsnr:+.2f} dB | SSIM={mssim:.3f} | "
            f"time={time.time()-eval_t0:.2f}s"
        )

    print(f"Epoch {epoch+1} finished in {time.time()-epoch_t0:.2f}s")

print(f"\nAll done. Total time: {time.time()-global_t0:.2f}s")
```
